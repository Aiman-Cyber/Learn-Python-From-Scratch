```
# Lets apply a while loop

i = 1

while i < 7:
  print(i)
  i = i+1      # We can also define this statement like this:  i+=1
```

```
# Using while loop with break statement

a = 1

while a < 10:
  print(a)
  if a == 5:
    break
  a +=1
```

```
# With the continue statement we can stop the current iteration, and continue with the next:

i = 0
while i < 10:
  i += 1
  if i == 4:
    continue
  print(i)
```
