```
# Lets Create a set in Python

a = {"Apple","Cherry","Mango","Pineapple"}
print(a)
print(type(a))
```

```
# Accessing items in a set
# We cannot access items in a set by referring to an index, since sets are unordered the items has no index.
# But still we can access items through a for loop.

a = {"Apple","Cherry","Mango","Pineapple"}

for item in a:
  print(item)
```

```
# Lets Check whether a particular element is present in a set or not.

a = {"Apple","Cherry","Mango","Pineapple"}
print("Apple" in a)
```

```
# Add a item in a set    # Single item     add()

a = {"Apple","Cherry","Mango","Pineapple"}
a.add("Kiwi")
print(a)
```

```
# Add Multiple Items in a set         # update()

a = {"Apple","Cherry","Mango","Pineapple"}
a.update(["Orange","Kiwi"])
print(a)
```

```
# Check the length of a set      len()

a = {"Apple","Cherry","Mango","Pineapple"}
print(len(a))
```

```
# Remove the item from a set      # remove() , discard() , pop()

a = {"Apple","Cherry","Mango","Pineapple"}
a.remove("Apple")
print(a)

b = {"Apple","Cherry","Mango","Pineapple"}
b.discard("Cherry")
print(b)

c = {"Apple","Cherry","Mango","Pineapple"}   # As we see pop() removes the last item but in this case sets are unordered and will not know what item that gets removed.
c.pop()
print(c)
```

```
# Lets Delete all elements from a set             # clear()

a = {"Apple","Cherry","Mango","Pineapple"}
a.clear()
print(a)
```

```
# Delete whole set      # del()

a = {"Apple","Cherry","Mango","Pineapple"}
del(a)
print(a)
```

```
# Joining two sets     # union()    # It returns a new set.
a = {"Apple","Cherry","Mango","Pineapple"}
b = {"Kiwi"}

c = a.union(b)
print(c)
```

```
# Update two sets     # update()

a = {"Apple","Cherry","Mango","Pineapple"}
b = {1,2,3}

a.update(b)
print(a)
```

<h3> Both Union and Update will Exclude any Duplicate Items.</h3>

```
# We can also create a set using constructor   set()

a = set(("Apple","Cherry","Mango","Pineapple"))
print(a)
print(type(a))
```

```
# Set difference()  method.
# Return a set that contains the items that only exist in set x, and not in set y.

x = {"A","B","C"}
y = {"A","E","I","O","U"}

z = x.difference(y)
print(z)
```

```
# Set difference_update()  method.
# Remove the items that exist in both sets.

m = {"A","B","C"}
n = {"A","E","I","O","U"}

m.difference_update(n)
print(m)
```

```
# interserction()   method.
# Return a set that contains the items that exist in both set x, and set y.

x = {"A","B","C"}
y = {"A","E","I","O","U"}
z = x.intersection(y) 
print(z)
```

```
# intersection_update()      method.
# Remove the items that is not present in both x, and set y.

x = {"A","B","C","E"}
y = {"A","E","I","O","U"}
x.intersection_update(y)
print(x)
```

```
# isdisjoint()    method.
# Return True if no items in set x is present in set y.

x = {"A","B","C","E"}      
y = {"A","E","I","O","U"}

z = x.isdisjoint(y)
print(z)
```

```
# issubset()    method
# Return True if all items set x are present in set y.

x = {"A","B","C"}      
y = {"A","b","C"}
z = x.issubset(y)
print(z)
```

```
# issuperset()     method
# Return True if all items set y are present in set x.

x = {"g", "h", "d", "c", "b", "a"}
y = {"a", "b", "c"}

z = x.issuperset(y)

print(z) 
```

```
# copy()  method
fruits = {"apple", "banana", "cherry"}

g = fruits.copy()

print(g) 
```

```
# symmetric_difference()     method.
# Return a set that contains all items from both sets, except items that are present in both sets.
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.symmetric_difference(y)

print(z) 
```

```
# symmetric_differece_update()     method.
# Remove the items that are present in both sets, AND insert the items that is not present in both sets.

x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.symmetric_difference_update(y)

print(x) 
```
