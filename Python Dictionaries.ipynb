<h3>Basic Syntax</h3>
<h4>{"Key" : "Value"}</h4>

```
# Lets Create a new Dictionary

a = {'1':"Pulkit",'2':"Rohan",'3':"Vaibhav",'4':"Tarun",'5':"Shivam"}
print(a)
print(type(a))
```

```
# Making a new dictionary through dict() constructor.

tdict = dict(name="Shayara",age='25',dob='15/11/1990')
print(tdict)
print(type(tdict))
```

```
# Accessing Names through Dictionary

x = a["1"]   # Here We access the name which is present in key 1
print(x)
```

```
y = a["3"]   # Here We access the name which is present in key 3
print(y)
```

```
# We can also get names through get()  function.
z = a.get("2")
print(z)
```

```
# Lets Create Another Dictionary

b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
print(b)
print(type(b))
```

```
# Now we want to change the value of key "a"
b["a"] = "Apricot"
print(b)
```

```
# Loop Through A Dictionary to get keys
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
for i in b:
  print(i)     # As u shown in this it gives key values
```

```
# Loop Through A Dictionary to get values
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
for i in b.values():
  print(i)     # As u shown in this it gives values
```

```
# Loop Through both keys and values
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}

for i,j in b.items():
  print(i,j)
```

```
# Check if key exists:
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}

if "o" in b:
  print("Yes")
else:
  print("No")
```

```
# Check if Value exists:
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}

if "Apple" in b.values():
  print("Yes")
else:
  print("No")
```

```
# Checking the Length of a dictionary
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
print(len(b))
```

```
# Add a new item in a dictionary
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}

b["g"] = "Grapes"
print(b)
```

```
# Remove a item from a dictionary
# The pop() method removes the item with the specified key name:
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
b.pop("a")
print(b)
```

```
# The popitem() method removes the last inserted item
b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
b.popitem()
print(b)
```

```
# The clear() method empties the dictionary:

b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
b.clear()
print(b)
```

```
# Making copy of a dictionary

b = {"a":"Apple","b":"Banana","m":"Mango","p":"Pineapple"}
m = b.copy()
print(m)
```

```
# Nested Dictionaries


job = {
  "emp1" : {
    "name" : "Kevin",
    "year" : 2019
  },
  "emp2" : {
    "name" : "Joseph",
    "year" : 2010
  },
  "emp3" : {
    "name" : "Kaif",
    "year" : 2014
  }
}

print(job)
```
