<h3>Python Supports Logical Conditions just like in Mathematics.</h3>


Equals: a == b <br>
Not Equals: a != b <br>
Less than: a < b <br>
Less than or equal to: a <= b <br>
Greater than: a > b  <br>
Greater than or equal to: a >= b <br>


```
# If - Else Statement:
```

<h3>Syntax</h3>

if condition:<br>
  &emsp;print(.....)<br>
else:<br>
  &emsp;print(.....)<br>

```
# Lets apply Equal Condition in if - else statements:

a = 50
b = 100

if a == b:
  print("A is Equal to B")
else:
  print("Both are Different")
```

```
# Not Equals a!=b

a = "Pulkit"
b = "Rohan"

if a!=b:
  print("Different")
else: 
  print("Same")
```

```
# Less then a < b

a = 10
b = 20

if a < b:
  print("A is Smaller")
else:
  print("B is Smaller")
```

```
# Less then equal to  a <= b

a = 10
b = 10

if a <= b:
  print("A is either Small or Equal")
else:
  print("B is Largest")
```

```
# Greater Than   a > b

a = 20
b = 30

if a > b:
  print("A is Greater")
else:
  print("A is Smaller")
```

```
# Greater Than equal to   a >= b

a = 40
b = 30

if a >= b:
  print("A is either Greater or equal to B")
else:
  print("A is Smaller")
```

<h3> Now Lets See elif keyword b/w if and else </h3>

elif keyword represents more conditions.

```
a = 100
b = 30

if b > a:                            # 1st Condition
  print("b is greater than a")     
elif a == b:                         # 2nd Condition
  print("a and b are equal")
else:                                # 3rd Condition
  print("a is greater than b")
```

<h3> Magical <B><I><U>AND</B></I></U> Keyword

```
# The and keyword is a logical operator, and is used to combine conditional statements.

a = 200
b = 33
c = 500
if a > b and a > c:
  print("A is Greater")
elif b > a and b > c:
  print("B is Greater")
else:
  print("C is Greater")
```

<h3> Magical <B><I><U>OR</B></I></U> Keyword

```
# The or keyword is a logical operator, and is used to combine conditional statements. Only 1 condition is required to be True.

a = 200
b = 33
c = 500
if a > b or a > c:
  print("One condition is True")   
```

<h3>Nested If Statements </h3>

```
number = 15

if number >= 0: 
    if number == 0: 
        print("Zero") 
    else: 
        print("Positive number") 
else: 
    print("Negative number") 
```

<h3><u>PASS</U> Keyword</h3>

if statements cannot be empty, but if you for some reason have an if statement with no content, put in the pass statement to avoid getting an error.

```
a = 10
b = 20

if a < b:
  pass
```
